# Price

가격이 낮은 애들 : dfm 증가(10%) lgb 감소(20%)

```python
dfm_list = []
lgb_list = []
X = range(50)
vi_df = pd.DataFrame(columns = {'lgb', 'dfm'})

for i in X:
    min_price = 50*10000 * i
    max_price = 50* 10000*(i+1)
    cond = (df['content_price'] > min_price) & (df['content_price']< max_price)
    # print(df[cond][['lgb_r', 'dfm_r']].mean())
    vi_df.loc[min_price] = {'lgb': df[cond]['lgb_r'].mean(), 'dfm': df[cond]['dfm_r'].mean()}
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=vi_df.index, y=vi_df.lgb, name="lgb",
                    line_shape='spline'))
fig.add_trace(go.Scatter(x=vi_df.index, y=vi_df.dfm, name="dfm",
                    line_shape='spline'))
fig.show()
```

# 키덜트

```python
df.groupby('content_cat_1').mean()[['dfm_r', 'is_ad']].sort_values('dfm_r', ascending=False).index[:5]
```

```python
ctr = ctr.set_index('content_id')
```

```python
for cat in df.groupby('content_cat_1').mean()[['dfm_r', 'is_ad']].sort_values('dfm_r', ascending=False).index:
    ctr_kid = ctr.loc[df[df['content_cat_1'] == cat].content_id].groupby(['gender', 'age']).mean()
    ctr_kid = ctr_kid.reset_index()
    ctr_kid['lgb'] = ctr_kid['lgb'].apply(lambda x : round(100*x/avg_lgb - 100, 2))
    ctr_kid['dfm'] = ctr_kid['dfm'].apply(lambda x : round(100*x/avg_dfm - 100, 2))
    print(f'== {cat} ==')
    fig = px.line(ctr_kid, x='age', y='dfm', color='gender')
    fig.show()
```

```python
vi_df = ctr_kid
fig = go.Figure()
fig.add_trace(go.Scatter(x=vi_df.index, y=vi_df.lgb, name="lgb", line_shape='spline'))
fig.add_trace(go.Scatter(x=vi_df.index, y=vi_df.dfm, name="dfm", line_shape='spline'))
fig.show()
```
